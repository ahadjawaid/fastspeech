# inference
> This module contains function used in inferences for using fastspeech models from checkpoints

```python
#| export
from fastspeech.modules import FastSpeech
from fastspeech.training import load_checkpoint
from fastspeech.visualize import *
from fastspeech.data import Vocab
import torch
import librosa
from fastspeech.preprocess import reduce_noise
from g2p_en import G2p
import re
```

```python
#| hide
from IPython.display import Audio
```

```python
#| hide
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
#| hide
checkpoint_path = "../checkpoints/save_590.pt"
vocab_path = "../sample_data/cmudict-0.7b.symbols.txt"
```

```python
#| export
def load_model_inference(Model, checkpoint_path, training=False):
    state_dict, config, norm = load_checkpoint(checkpoint_path, 'cpu')
    model = Model(config["model"])
    model.load_state_dict(state_dict['model'])
    model = model.train(mode=training)
    return model, norm
```

```python
model, norm = load_model_inference(FastSpeech, checkpoint_path)
```

```python
#| hide
text = "This is a demonstration of my implementation of the text to speech model based on the fast speech's research paper using the griffin-lim algorithm as the vocoder"
```

```python
#| export
def filter_non_alpha_chars(input_str):
    return re.sub(r'[^A-Za-z ]', '', input_str)
```

```python
filter_non_alpha_chars(text)
```

```python
#| export
g2p = G2p()
```

```python
#| export
def preprocess_text(text, vocab_path):
    vocab = Vocab(vocab_path)
    text = filter_non_alpha_chars(text)
    phones = g2p(text)
    phones = list(filter(lambda x: x != ' ', phones))
    vec = torch.tensor(list(map(vocab.tok2idx.__getitem__, phones))).unsqueeze(0)
    return vec
```

```python
vec = preprocess_text(text, vocab_path)
vec.shape
```

```python
#| export
def inference(vec, model, upsample_ratio=1.):
    model = model.train(mode=False)
    with torch.no_grad(): mel = model(vec, upsample_ratio=upsample_ratio)
    return mel.squeeze()
```

```python
mel = norm.denormalize(inference(vec, model))
show_mel(mel)
```

```python
#| export
def bayesian_inference(vec, model, samples=10, upsample_ratio=1.):
    '''Does ensemabling based on averaging between different 
    prediction with dropout left on to add varience between each predicition'''
    model = model.train(mode=True)
    
    durr_preds = torch.stack([model(vec)[1] for _ in range(samples)])
    durr_mean = durr_preds.mean(dim=0)
    durr_stacked = torch.stack([durr_mean.squeeze() for d in durr_preds])
    vec_stacked = torch.stack([vec.squeeze() for _ in range(samples)])

    with torch.no_grad(): mels = model(vec_stacked, durr_stacked, upsample_ratio)[0]
    averaged_mel = mels.mean(dim=0)
    return averaged_mel
```

```python
mel_b = norm.denormalize(bayesian_inference(vec, model, 10))
show_mel(mel)
```

```python
#| hide
wav = librosa.feature.inverse.mel_to_audio(mel.numpy(), sr=22050, n_fft=1024, 
                                           hop_length=256, n_iter=16)
wav_b = librosa.feature.inverse.mel_to_audio(mel_b.numpy(), sr=22050, n_fft=1024, 
                                           hop_length=256, n_iter=16)
```
